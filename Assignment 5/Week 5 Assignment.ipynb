```python

# read the csv into a dataframe
weather = pd.read_csv('weather.csv')

# Display the columns in the dataframe, and get a peak at how it's formatted
weather.head()
```

```python
# Filter out the correct year, month and day.
# Then sort the dataframe by wind_speed
# Using this it would appear that EWR had the windiest weather on 12Feb13. However the windspeed on that
# day appears to be incorrect. 
weather[((weather.year == 2013) & \
         (weather.month == 2)) & \
        (weather.day == 12)] \
        .sort_values('wind_speed', ascending = False)
```

```python

# Filter out the correct year, month and day. Also filter out incorrect windspeeds I used 350 because 
# 301 ± 20 mph were the highest winds ever recorded on earth (https://en.wikipedia.org/wiki/Tornado_records)
# I Then sort the dataframe by wind_speed
# Using this it would appear that LGA had the windiest weather on 12Feb13. However the this dataset is broken
# down by hour.
weather[((weather.year == 2013) & \
         (weather.month == 2)) & \
        (weather.day == 12) & \
        (weather.wind_speed < 350)] \
        .sort_values('wind_speed', ascending = False)
```

```python
# To account for the fact that the wind data is taken every hour I will take the average hourly 
# windspeed for each airport that day.
weather[((weather.year == 2013) & \
         (weather.month == 2)) & \
        (weather.day == 12) & \
        (weather.wind_speed < 350)] \
        .groupby('origin').wind_speed.mean()
```
